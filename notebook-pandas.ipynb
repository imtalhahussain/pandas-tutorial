```python
# To apply filter to show the data 
df[df["Category"]=="Electronics"]
```

```python
# If i want to apply multiple filters then for AND
df[(df["Category"]=="Electronics") & (df["Country"]=="USA")]
```

```python
# If i want to apply multiple filters then for OR
df[(df["Category"]=="Electronics") | (df["Country"]=="USA")]
```

```python
# If I want to search name using Alphabet Which starts with
df[df["CustomerName"].str.startswith("A")]
```

```python
# If I want to search name using Alphabet Which ends with
df[df["CustomerName"].str.endswith("a")]
```

```python
df[df["Country"].isin(["USA","Sweden"])]
```

```python
df[~df["Country"].isin(["USA"])]
```

```python
#Updating Data - For updating a product using customer Name
df.loc[df["CustomerName"]=="Daniel Kim","Product"] ="Tim"
```

```python
#DELETE Data
df = df.drop(39)
```

```python
#Cleaning Data - Removes Null values
df.dropna() 
```

```python
#   Analyzing Data - if i want to see how many products are from USA 
df["Country"].value_counts()
```

```python
# Sums up the value by grouping
df.groupby("Country")["Price"].sum()
```

```python
df.sort_values("Price" , ascending=False)
```
